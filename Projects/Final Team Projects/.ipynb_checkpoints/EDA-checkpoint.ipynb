### FOMC EDA

#### FOMC Speech

```python
import numpy as np
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import DataLoader, TensorDataset
import nltk
from nltk.corpus import stopwords

model_name = "yiyanghkust/finbert-tone"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning) # FutureWarning 제거

speech = pd.read_excel('../../Data/Final/speech_df.xlsx', index_col = 1)
speech.index = pd.to_datetime(speech.index)
unique_dates = speech.index.unique()
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

text_for_date = speech["speech"].tolist()

tokenized_texts = [
        [word for word in tokenizer.tokenize(text) if word.lower() not in stop_words]
        for text in text_for_date]
```

```python
len_text = []
for i in tokenized_texts : len_text.append(len(i))
```

```python
len_df = pd.DataFrame(
    {'year': speech.index.strftime('%Y'),
     'corpus_length' : len_text,},
    index = speech.index
)
len_df.sort_index(ascending = True, inplace = True)
```

```python
len_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 4))
plt.grid(False)
plt.scatter(len_df.index, len_df['corpus_length'], marker = '.', color = 'navy')
```

```python
import seaborn as sns

plt.figure(figsize = (8, 4))
plt.axvspan(-1, 8, color = 'orange', alpha = 0.2)
plt.axvspan(8, 12, color = 'green', alpha = 0.2)
plt.axvspan(12, 18, color = 'purple', alpha = 0.2)
sns.boxplot(
    x = 'year', 
    y = 'corpus_length', 
    data = len_df,
    palette = ['darkorange']
)
plt.text(4, 0.8, 'Bernanke', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.text(10, 0.8, 'Yellen', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.text(15, 0.8, 'Powell', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.grid(False)
plt.title('FOMC Speech corpus length')
plt.show()
```

```python
import time 
start_time = time.time()

daily_sentiments = pd.DataFrame(columns = ["Date", "Positive", "Negative", "Neutral"])

nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

for date in unique_dates:    
    #text_for_date = data[data["rgs_dt"] == date]["news_smy_ifo"].tolist()
    text_for_date = speech[speech.index == date]["speech"].tolist()

    tokenized_texts = [
        [word for word in tokenizer.tokenize(text) if word.lower() not in stop_words]
        for text in text_for_date
    ]

    max_token_length = 512

    tokenized_texts = [" ".join(tokens[:max_token_length]) for tokens in tokenized_texts]
    inputs = tokenizer(tokenized_texts, padding = True, truncation = True, return_tensors = "pt", max_length = max_token_length)

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits

    probabilities = torch.softmax(logits, dim=1)
    positive_prob = probabilities[:, 2].mean().item()
    negative_prob = probabilities[:, 0].mean().item()
    neutral_prob = probabilities[:, 1].mean().item()
    
    daily_sentiments = pd.concat([daily_sentiments, 
                                  pd.DataFrame({"Date": date, 
                                                "Positive": positive_prob, 
                                                "Negative": negative_prob, 
                                                "Neutral": neutral_prob}, index = ['0'])],
                                 ignore_index = True)

print(f'spent time : {time.time() - start_time}')
daily_sentiments.index = daily_sentiments.Date
daily_sentiments = daily_sentiments.iloc[:,1:]
daily_sentiments
```

```python
sentiment_data = daily_sentiments.resample('M').mean()
sentiment_data['year'] = sentiment_data.index.strftime('%Y')
```

```python
sentiment_data
```

```python
import seaborn as sns

plt.figure(figsize = (8, 4))
plt.axvspan(-1, 8, color = 'orange', alpha = 0.2)
plt.axvspan(8, 12, color = 'green', alpha = 0.2)
plt.axvspan(12, 18, color = 'purple', alpha = 0.2)
sns.boxplot(
    x = 'year', 
    y = 'Positive', 
    data = sentiment_data,
    palette = ['darkorange']
)
plt.text(4, 0.85, 'Bernanke', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.text(10, 0.85, 'Yellen', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.text(15, 0.85, 'Powell', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.grid(False)
plt.title('FOMC Speech positive sentiment rate')
plt.show()
```

```python
sentiment_data['sentimental_score'] = (sentiment_data['Positive'] - sentiment_data['Negative']) + sentiment_data['Neutral']*0.5
```

#### FOMC Press 

```python
press = pd.read_excel('../../Data/Final/press_df.xlsx', index_col = 1)
press.index = pd.to_datetime(press.index)
unique_dates = press.index.unique()
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
text_for_date = press["speech"].tolist()
tokenized_texts = []
for text in text_for_date :
    try : 
        temp = []
        for word in tokenizer.tokenize(text) :
            if word.lower() not in stop_words :
                temp.append(word)
        tokenized_texts.append(temp)
    except : 
        tokenized_texts.append([])
```

```python
len_text = []
for i in tokenized_texts : len_text.append(len(i))
```

```python
len_df = pd.DataFrame(
    {'year': press.index.strftime('%Y'),
     'corpus_length' : len_text,},
    index = press.index
)
len_df.sort_index(ascending = True, inplace = True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 4))
plt.grid(False)
plt.scatter(len_df.index, len_df['corpus_length'], marker = '.', color = 'navy')
```

```python
import seaborn as sns

plt.figure(figsize = (8, 4))
plt.axvspan(-1, 8, color = 'orange', alpha = 0.2)
plt.axvspan(8, 12, color = 'green', alpha = 0.2)
plt.axvspan(12, 18, color = 'purple', alpha = 0.2)
sns.boxplot(
    x = 'year', 
    y = 'corpus_length', 
    data = len_df,
    palette = ['darkorange']
)
plt.text(4, 1500, 'Bernanke', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.text(10, 1500, 'Yellen', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.text(15, 1500, 'Powell', 
         horizontalalignment = 'center', 
         verticalalignment = 'center')
plt.grid(False)
plt.title('FOMC Press corpus length')
plt.show()
```

```python
import time
daily_sentiments_excerpt = pd.DataFrame(columns = ["Date", "Positive", "Negative", "Neutral"])
start_time = time.time()

nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

for date in unique_dates:    
    #text_for_date = data[data["rgs_dt"] == date]["news_smy_ifo"].tolist()
    text_for_date = press[press.index == date]["speech"].tolist()

    tokenized_texts = []
    for text in text_for_date :
        try : 
            temp = []
            for word in tokenizer.tokenize(text) :
                if word.lower() not in stop_words :
                    temp.append(word)
            tokenized_texts.append(temp)
        except : 
            tokenized_texts.append([])

    max_token_length = 512

    tokenized_texts = [" ".join(tokens[:max_token_length]) for tokens in tokenized_texts]
    inputs = tokenizer(tokenized_texts, padding = True, truncation = True, return_tensors = "pt", max_length = max_token_length)

    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits

    probabilities = torch.softmax(logits, dim=1)
    positive_prob = probabilities[:, 2].mean().item()
    negative_prob = probabilities[:, 0].mean().item()
    neutral_prob = probabilities[:, 1].mean().item()
    
    daily_sentiments_excerpt = pd.concat([daily_sentiments_excerpt, 
                                          pd.DataFrame({"Date": date, 
                                                        "Positive": positive_prob, 
                                                        "Negative": negative_prob, 
                                                        "Neutral": neutral_prob}, index = ['0'])],
                                 ignore_index = True)

daily_sentiments_excerpt.index = daily_sentiments_excerpt.Date
daily_sentiments_excerpt = daily_sentiments_excerpt.iloc[:,1:]
daily_sentiments_excerpt
```
